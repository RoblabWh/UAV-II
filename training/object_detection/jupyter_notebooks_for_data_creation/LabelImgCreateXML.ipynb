### Creating new XML files for multiple backgrounds and saving them

```python
## CREATING FILES WITH COLLECTED TREES FROM ABOVE

# laserSign, radiationSign
objName = "laserSign"
path = "/home/tiffany/master_projekt/dataset_collection/laserSign/combined_1/background_"

for k in range (9,10):
    background_path = "/home/tiffany/master_projekt/dataset_collection/laserSign/combined_"+str(k)+"/"
    for i in range (len(tree_list)):
        background_folder = "background_"+str(i+1)+"/"
        root = tree_list[i].getroot()
        for file in os.listdir(background_path+background_folder):
            img = cv2.imread(background_path+background_folder+file)
            height, width, depth = img.shape
            root.find("./filename").text = file
            root.find(".//name").text = objName
            root.find(".//height").text = str(height) 
            root.find(".//width").text = str(width)
            root.find(".//depth").text = str(depth)
            tree_list[i].write(background_path+background_folder+os.path.splitext(file)[0]+".xml")
os.system('spd-say "XML files for '+objName+' were created"')
```
