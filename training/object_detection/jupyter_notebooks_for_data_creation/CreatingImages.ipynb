## Create Images

```python
rotated = dc.rotateImage(img_path, img_name, 7)
print("rotated size:", len(rotated))
dc.saveImages("rotated", img_path+curr_directory, rotated)
```

```python
red_img = []
gray_img = []
invert_img = []
blur = []
blur_kernels = [3,5,7,9] # 3,5,7,9,13 normally 
endings = (".jpg", ".png")
for file in os.listdir(img_path+curr_directory):
    if file.endswith(endings):
        red_img.append(dc.colorImageRed(img_path+curr_directory, file))
        gray_img.append(dc.colorImageGray(img_path+curr_directory, file))
        invert_img.append(dc.invertImageColor(img_path+curr_directory, file))
        for i in range(len(blur_kernels)):
            blur.append(dc.gaussianFilter(img_path+curr_directory, file, blur_kernels[i]))

print("Blur", len(blur))
print("Red:", len(red_img))
print("Gray:", len(gray_img))
print("Invert:", len(invert_img))
```
